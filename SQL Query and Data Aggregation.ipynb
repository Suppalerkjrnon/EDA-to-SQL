# SQL Query

```python
# Connect to the database
db_path = r'F:\For work\Database\kaggle_movie.db'

conn = sqlite3.connect(db_path)

#Query the database to get all data from table 'kaggle_movie'
query = "SELECT * FROM kaggle_movie"

kaggle_movie_df = pd.read_sql(query, conn)

conn.close()
```

## Query using Where statement

```python
conn = sqlite3.connect(db_path)

#Query the database to get all data from table 'kaggle_movie'
query = "SELECT * FROM kaggle_movie Where year >= '2016'"

kaggle_movie_2016 = pd.read_sql(query, conn)

conn.close()
```

```python
kaggle_movie_2016
```

## Query Unique record

```python
conn = sqlite3.connect(db_path)

#Query the database to get all data from table 'kaggle_movie'
query = "SELECT DISTINCT movies FROM kaggle_movie"

kaggle_unique_movie = pd.read_sql(query, conn)

conn.close()
```
